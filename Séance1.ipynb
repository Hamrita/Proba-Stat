# Expression de la fonction $F(x)$:
Par définition
$$F(x)=\int_{-\infty}^xf(t)dt
$$
- Si $x <0$: $F(x)=\int_{-\infty}^x 0 \times dt=0$.
- Si $0 \leq x \leq 1$; $F(x)=\int_{-\infty}^00\times dt+\int_{0}^x 1\times dt=x$.
- Si $x > 1$: $F(x)=\int_{-\infty}^0 0\times dt+\int_0^1 1\times dt+\int_1^x 0\times dt=1$

$$
F(x)=\begin{cases}
0 \text{ si } x <0\\
x \text{ si } x \in [0,1]\\
1 \text{ si }  x > 1
\end{cases}
$$

$$
